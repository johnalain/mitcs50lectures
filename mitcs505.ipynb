```python
#convert to binary num 4
x = 0.625
#grabs the decimal part only e.g 1.1%1 gives o.1
p = 0
while ((2**p)*x)%1 !=0:
	print("remainder = " + str((2**p)*x -int(2**p)*x))
	p += 1
#convert to int
num = int(x*(2**p))
#encode as binary number same as previous slides
result = ''
if num == 0:
	result = '0'
while num > 0:
	result = str(num%2) +result
	num = num//2
	#pad front with 0's e.g shift right
for i in range(p -len (result)):
	result = '0' +result
	#insert decimal
result = result[0:-p]+'.'+ result[-p:]
print("the binary representation of the decimal " + str(x)+' is '+ str(result))
```

```python
# approximation
x = 54321
epsilom = 0.01
num_guesses = 0
guess = 0.0
increment = 0.00001
while abs(guess**2 - x) >= epsilom:
        guess += increment
        num_guesses += 1
print(f"num_guesses = {num_guesses}")
if abs(guess**2 - x)>=epsilom:
	print(f"failed on square root of {x}")
	print(f"last guess was {guess}")
	print(f"last guess squared is {guess*guess}")
else:
		print(f'{guess} is close to square root of {x}')
```
