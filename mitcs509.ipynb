```python
def apply(criteria, n):
        count = 0
        for i in range(n+1):
            if criteria(i):
                    count+=1
        return count
def is_even(x):
        return x%2 == 0
def is_five(x):
       return x == 5
print('apply with is_five',apply(is_five,10))
print('apply with ano fcn',apply(lambda x: x ==5,10))
print((lambda x: x%2==0)(5))
```


seq =(2,'a',4,(1,2))what is the output of print(seq[1:4:2])

Let's break it down:

seq = (2, 'a', 4, (1, 2))
print(seq[1:4:2])
Step 1: Understand slicing
The format is:

seq[start : stop : step]
start = 1 → start from index 1 → 'a'

stop = 4 → go up to (but NOT including) index 4

step = 2 → take every 2nd element

Step 2: Indices of the tuple
Index:   0    1    2     3
Value:   2   'a'   4   (1,2)
Step 3: Apply slicing
Start at index 1:

index 1 → 'a'

move +2 → index 3 → (1, 2)

Final result:
('a', (1, 2))
✅ Output:

('a', (1, 2))
If you want, I can show a visual way to understand slicing faster.







```python
seq =(2,'a',4,(1,2))
print(seq[-2])# 4
print(seq[1:4:2])#
print(seq[:-1])
print(seq[1:3])
for e in seq:
        print(e)
```

```python
#  usually function return one return but with tuple vcan return multiple values
def quotient_and_remainder(x,y):
        q = x//y
        r = x%y
        return (q,r)
both = quotient_and_remainder(10,3)
print(both)
(quot,rem) = quotient_and_remainder(5,2)
print((quot,rem))
```

```python
def char_counts(s):
        vowels = 'aeiou'
        (c,v) =(0,0)
        for char in s:
            if char in vowels:
                v+=1
            else:
                c+=1
        return (v,c)
print(char_counts('abcd'))
print(char_counts('zcght'))
```

```python
def mean(*args):
    tot = 0
    for a in args:
        tot += a
    return tot/len(args)
x = mean(1,5,6,4,3,7,8,9)
print(x)
```
