```python
x = 36
epsilon = 1
num_guesses = 0
#initializesome stuff
low = 0
high =x
guess =(high+low)/2.0

while abs(guess**2 -x) >= epsilon:
# what si repeated each time?
	if guess**2 <= x:
		low = guess
	else:
		high = guess
	guess = (high + low)/2.0
	num_guesses +=1
print('number of gusses = ',num_guesses)
print(guess,'is close to square root of ',x)

```
