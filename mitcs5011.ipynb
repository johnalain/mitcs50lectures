```python
import copy


old_list =[[1,2],[3,4],[5,'foo']]
new_list = copy.copy(old_list)
# copy() mutates both new_list and old_list
old_list.append([7, 8])
old_list[1][1]=9
print("New list:  ",new_list)
print("old list:  ",old_list)
```

```python
# use deepcopy
import copy
old_list =[[1,2],[3,4],[5,'foo']]
new_list = copy.deepcopy(old_list)
# deepcopy() mutates only old_list

old_list.append([7, 8])
old_list[1][1]=9
print("New list:  ",new_list)
print("old list:  ",old_list)
```
