Black-box testing in Python is a testing method where you check a function or program only by its inputs and outputs, without looking at how the code works internally.

🧠 Simple Idea
Think of your function like a “black box”:

You don’t care how it works inside

You just:

Give it input

Check if the output is correct

✅ Example
Function to test
def add(a, b):
    return a + b
Black-box tests
def test_add():
    assert add(2, 3) == 5
    assert add(-1, 1) == 0
    assert add(0, 0) == 0
👉 You are NOT checking:

variables inside the function

how + works

👉 You ARE checking:

“Does it return the correct result?”

🧪 Using pytest (popular tool)
Install:

pip install pytest
Create file: test_mycode.py

from mycode import add

def test_add():
    assert add(2, 3) == 5
Run:

pytest
🎯 What to Test (Black-box mindset)
Focus on:

✔️ Normal cases

✔️ Edge cases

✔️ Invalid input

Example
def divide(a, b):
    return a / b
Tests:

def test_divide():
    assert divide(10, 2) == 5
    assert divide(5, 1) == 5

def test_divide_by_zero():
    try:
        divide(5, 0)
        assert False  # should not reach here
    except ZeroDivisionError:
        assert True
🔥 Key Difference
Type	What you test
Black-box	Input → Output
White-box	Internal code logic
💡 Real-life analogy
ATM machine:

You insert card + PIN

You get money (or error)

You don’t know the internal system → that’s black-box testing

🚀 Summary
Test behavior, not implementation

Focus on inputs and outputs

Use tools like pytest

Cover edge cases

glass box testing

Glass-box testing (also called white-box testing or structural testing) is a software testing method where you test the internal logic and structure of the code.

🔍 What it means
In glass-box testing, the tester:

Can see the source code

Understands how the program works internally

Writes tests based on code paths, conditions, loops, and logic

Think of it like testing a machine with a transparent (glass) cover — you can see everything inside.

🧠 Key idea
Instead of only checking inputs and outputs (like in black-box testing), you test:

How the code executes

Which lines are run

Whether all conditions are covered

🧪 Example in Python
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False
Glass-box testing approach:
You look at the code and test both branches:

print(is_even(4))  # tests True branch
print(is_even(3))  # tests False branch
Here, you are making sure:

The if condition is tested

The else condition is tested

🧩 Types of glass-box testing
Statement Coverage
→ Make sure every line of code runs at least once

Branch Coverage
→ Test all possible paths (if/else)

Path Coverage
→ Test all possible execution paths (more detailed)

Loop Testing
→ Test loops (0 times, 1 time, many times)

⚖️ Glass-box vs Black-box
Feature	Glass-box	Black-box
Code visibility	✅ Yes	❌ No
Focus	Internal logic	Output behavior
Who uses it	Developers	Testers/users
✅ Advantages
Finds hidden bugs in logic

Ensures high code cover

how to use rubberducking

Rubber duck debugging (often called “rubber ducking”) is a simple but powerful way to debug your code by explaining it out loud—usually to an object like a rubber duck 🐤 (or anything really).

🧠 How it works
When you explain your code step-by-step, your brain is forced to slow down and think clearly. This often helps you spot mistakes you didn’t notice before.

✅ How to use rubber ducking (step by step)
1. Pick your “duck”
A real rubber duck, a pen, your screen, or even an imaginary listener

The key: treat it like it understands nothing

2. Explain your goal
Say out loud:

“This function should add two numbers and return the result.”

3. Walk through your code line by line
Example:

def add(x, y):
    return x + y
You would say:

“This defines a function called add”

“It takes x and y”

“It returns x plus y”

4. Be VERY detailed
Even obvious things:

“This loop runs from 0 to 10”

“This condition checks if the number is even”

👉 This is where bugs reveal themselves.

5. Ask yourself questions
“Wait… what happens if x is negative?”

“Why am I returning inside the loop?”

“Is this variable updated correctly?”

6. Find the bug
Most of the time, you’ll suddenly realize:

“Oh… that line is wrong.”

🔥 Example (your kind of bug)
You had something like:

for i in range(n):
    if condition:
        return count
Rubber ducking would make you say:

“Wait… I’m returning inside the loop… so it stops early…”

💡 That’s the bug!

🎯 Why it works
Forces clear thinking

Catches logical errors

Improves understanding

No tools needed
